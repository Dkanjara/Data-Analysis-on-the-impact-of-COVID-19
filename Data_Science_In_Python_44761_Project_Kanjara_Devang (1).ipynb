# Data Cleaning
* We will eliminate any useless data from the dataframe in this stage and only keep the important information.

```python
## obtaining and adding to the list the total number of columns in the data
total_columns = list(df.columns)   
total_columns
```

```python
## We will now establish a new dataframe with just the crucial and useful columns
useful_data = df[['location','date','new_cases','new_deaths','icu_patients','hosp_patients','total_tests','new_tests','positive_rate',
                  'total_vaccinations','people_vaccinated','people_fully_vaccinated','stringency_index','aged_65_older','aged_70_older',
                  'cardiovasc_death_rate','diabetes_prevalence','female_smokers','male_smokers', 'population']]
```

```python
## Adding one column in our dataframe
useful_data['life_expectancy'] = df.life_expectancy
useful_data.head()
```

```python
## Out of the total_columns in the "owid_covid_data," these are the sorted columns.
## We will only evaluate the data using these columns as these are the only ones that are helpful.
useful_columns = useful_data.columns
useful_columns
```

```python
## counting how many NULL values are there in our helpful dataset
useful_data.isnull().sum()
```

##### * As a result of the sheer number of NULL values in our dataset, we must fill these null values with acceptable values for certain columns before doing any data analysis.

```python
useful_data.head()
```

##### A) putting a '0' in each of the NULL values in the following columns
1. new_cases
2. new_deaths
3. icu_patients
4. hosp_patients
5. total_tests
6. new_tests
7. total_vaccinations
8. people_vaccinated
9. people_fully_vaccinated
10. female_smokers
11. male_smokers

```python
useful_data[['new_cases','new_deaths','icu_patients','hosp_patients','total_tests','new_tests','total_vaccinations','people_vaccinated','people_fully_vaccinated','female_smokers','male_smokers']] = useful_data[['new_cases','new_deaths','icu_patients','hosp_patients','total_tests','new_tests','total_vaccinations','people_vaccinated','people_fully_vaccinated','female_smokers','male_smokers']].fillna(value=0)
```

```python
## Check the null values once again. 
## We can observe that there are no NULL values in the aforementioned columns.
useful_data.isnull().sum()
```

##### B) 'Median' will be filled in for any NULL values in the columns listed below.
1. positive_rate
2. stringency_index
3. aged_65_older
4. aged_70_older

```python
## We will first get the median values for each column
e = useful_data['positive_rate'].median()
f = useful_data['stringency_index'].median()
g = useful_data['aged_65_older'].median()
h = useful_data['aged_70_older'].median()
```

```python
## We will now establish a dictionary where the keys are the names of the columns and the values are the median values for those columns
median_values = {'positive_rate':e, 'stringency_index':f,'aged_65_older':g,'aged_70_older':h}
```

```python
## We will now use the fillna function and input the dictionary as a value.
useful_data[['positive_rate','stringency_index','aged_65_older','aged_70_older']] = useful_data[['positive_rate','stringency_index','aged_65_older','aged_70_older']].fillna(median_values)
```

##### c) putting 'Mean' in the columns below when NULL values are present
1. cardiovasc_death_rate
2. diabetes_prevalence
3. life_expectancy

```python
## We will first determine the mean values for each column.
c = useful_data['cardiovasc_death_rate'].mean()
d = useful_data['diabetes_prevalence'].mean()
e = useful_data['life_expectancy'].mean()
```

```python
## We will now construct a dictionary where the keys are the names of the columns and the values are the means of the corresponding columns.
mean_values = {'cardiovasc_death_rate':c,'diabetes_prevalence':d,'life_expectancy':e}
```

```python
## We will now use the fillna function and add the dictionary as a value.
useful_data[['cardiovasc_death_rate','diabetes_prevalence','life_expectancy']] = useful_data[['cardiovasc_death_rate','diabetes_prevalence','life_expectancy']].fillna(mean_values)
```

```python
useful_data.isnull().sum()
```

##### As we can see, there are only 1071 null entries in the "population" column, which is a relatively tiny number when compared to the total amount of data.
##### Therefore, we will remove all of the remaining null values since they are not very significant and won't pose any problems with our dataframe.

```python
## before eliminating null values, the number of rows
np.shape(useful_data)
```

```python
## Number of rows after dropping null values
np.shape(useful_data)
```

# Data Analyzation
* This stage will include analyzing our usable data, producing some interesting dataframes, and creating various graphs.

```python
## gathering the total number of cases by grouping the dataframe by locales
total_cases = useful_data.groupby('location')['new_cases'].sum()
## sorting from highest to lowest 
total_cases = total_cases.sort_values(ascending=False)   

## picking five nations at random
most_affected_countries = total_cases[9:14].index
most_affected_countries
```

```python
## Fetching number of Cases in South America, India, France, Germany, Brazil
cases = total_cases[9:14].values 
print(cases)
```

```python
## gathering the overall death toll by grouping the dataframe by places
total_deaths = useful_data.groupby('location')['new_deaths'].sum()
deaths = total_deaths[most_affected_countries].values # Number of Deaths in South America, India, France, Germany, Brazil
print(deaths)
```

```python
## creating a dataframe with information on the number of cases and deaths in each of five nations
cases_deaths_df = pd.DataFrame({'Country': most_affected_countries, 'Total Cases': cases, 'Total Deaths': deaths})
cases_deaths_df
```

```python
## Establishing a vertical bar graph of Country vs. Total Cases
sns.barplot(x = 'Country', y = 'Total Cases', data=cases_deaths_df, palette = 'crest')

plt.title('Total number of Cases')
```

##### According to the above graphic, South America has the most instances among these five nations.

```python
## Establishing a vertical bar graph of Country vs. Total Deaths
sns.barplot(x = 'Country', y = 'Total Deaths', data=cases_deaths_df, palette = 'flare')
    
plt.title('Total number of Deaths')
```

##### According to the above graphic, South America has the most deaths among these five nations.

##### We will continue and do more study now that we have discovered the relationship between the total cases and deaths in certain nations.

```python
## determining the frequency of testing in various countries
total_tests = df.groupby('location')['new_tests'].sum().values
population = df.groupby('location')['population'].nth(-1)   # nth(-1) = the last value of population of particular country
testing_rate = (total_tests/population).sort_values(ascending = False)[:15]  # highest to lowest
```

```python
## the construction of a dataframe with the columns "Country" and "Testing_Rate"
testing_rate_df = pd.DataFrame({'Country':testing_rate.index,'testing_rate(total_tests/population)':testing_rate.values})
testing_rate_df
```

```python
## Plotting a horizontal bar plot
sns.barplot(x = testing_rate.values, y = testing_rate.index, palette="vlag")
plt.title('Testing Rate (Top 15 Countries)')
plt.xlabel('Total Tests / Total Population')
plt.ylabel('Country')
```

##### Out of all the nations, Cyprus has the highest testing rate.
